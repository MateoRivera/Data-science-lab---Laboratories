```python
index_to_drop = [] 

for i, invoice in enumerate(UCI['InvoiceNo']): #index, value
    if invoice[0] == 'C':
        index_to_drop.append(i)

UCI = UCI.drop(index_to_drop)  # drop the rows

print(UCI)
```

```python
UCI['Description']=UCI['Description'].fillna('Without register')
```

```python
UCI['CustomerID']=UCI['CustomerID'].fillna('Without register')
```

```python
keys=UCI['InvoiceNo']
```

```python
tupla=tuple(keys)
```

```python
values=UCI['StockCode'].values
```

```python
dict={tupla:values}
```

```python
invoice_order_dict = {}

for index, row in UCI.iterrows():
    invoice_no = row['InvoiceNo']
    order = row['Order']

    if invoice_no in invoice_order_dict:
        invoice_order_dict[invoice_no].append(order)
    else:
        invoice_order_dict[invoice_no] = [order]

# Imprimir el diccionario resultante
print(invoice_order_dict)
```
